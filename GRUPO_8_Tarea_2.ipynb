### Problema 2

```python
data_ciudades_2 = {
   'Ciudad': ['Lima', 'Arequipa', 'Trujillo', 'Cusco', 'Chiclayo', 'Piura', 'Iquitos', 'Huancayo', 'Tacna', 'Pucallpa'],
    'Gentilicio': ['Limense', 'Arequipeño', 'Trujillano', 'Cusqueño', 'Chiclayano', 'Piurano', 'Iquiteño', 'Huancaino', 'Tacneño', 'Pucallpino'],
    'Provincia': ['Lima', 'Arequipa', 'Trujillo', 'Cusco', 'Chiclayo', 'Piura', 'Maynas', 'Huancayo', 'Tacna', 'Coronel Portillo'],
    'Region': ['Lima', 'Arequipa', 'La Libertad', 'Cusco', 'Lambayeque', 'Piura', 'Loreto', 'Junín', 'Tacna', 'Ucayali']
   }

nombres = pd.DataFrame(data_ciudades_2)
nombres

```

### Problema 3

```python
cuadro_1=pd.merge(nfo, nombres, on='Ciudad', how='inner')
cuadro_1
```

### Problema 4

```python
cuadro_1.describe()
```

```python
min_densidad = cuadro_1['Densidad_poblacion_por_km2'].min()
max_densidad = cuadro_1['Densidad_poblacion_por_km2'].max()
print("La mínima densidad es", min_densidad)
print("La máxima densidad es", max_densidad)
```

```python

indice_max_densidad = cuadro_1['Densidad_poblacion_por_km2'].idxmax()
indice_min_densidad = cuadro_1['Densidad_poblacion_por_km2'].idxmin()

ciudad_max_densidad = cuadro_1.loc[indice_max_densidad, 'Ciudad']
ciudad_min_densidad = cuadro_1.loc[indice_min_densidad, 'Ciudad']

print("Ciudad con la máxima densidad poblacional:", ciudad_max_densidad)
print("Ciudad con la mínima densidad poblacional:", ciudad_min_densidad)
```

### Problema 5
